```python
import cv2
import numpy as np
from cedar.draw import color_list,imshow
from cedar.image import merge_boxes
```

```python
# 示例边界框列表
boxes = [
    [100, 100, 50, 50],  # [x, y, w, h]
    [120, 110, 45, 45],
    [200, 200, 60, 60]
]

# 创建一个空白图像
blank_image = np.zeros((400, 400, 3), dtype=np.uint8)
# 绘制保留的边界框
for bbox in boxes:
    x_min, y_min, w, h = bbox
    x_max,y_max = x_min+w, y_min+h
    cv2.rectangle(blank_image, (x_min, y_min), (x_max, y_max),color_list[0], 3)
# imshow(blank_image)

# 合并相近的边界框
merged_boxes = merge_boxes(boxes, iou_threshold=0.2)

print("合并后的边界框列表:", merged_boxes)
# 创建一个空白图像
# blank_image = np.zeros((400, 400, 3), dtype=np.uint8)
# 绘制保留的边界框
for bbox in merged_boxes:
    x_min, y_min, w, h = bbox
    x_max,y_max = x_min+w, y_min+h
    cv2.rectangle(blank_image, (x_min, y_min), (x_max, y_max),color_list[1], 1)
imshow(blank_image)
```
